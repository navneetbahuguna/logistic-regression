# 1) Import the required libraries (sklearn, pandas, numpy). Import datasets. Concatenate the datasets. Display first 5 rows.

```python
data_train=pd.read_csv("tit_train.csv")
data_train.head()
```

```python
data_test=pd.read_csv("tit_test.csv")
data_test.head()
```

# 2) Remove NaN values from the column ‘Survived’. 

```python
data_train.count()
# we can easily that only Age,Cabin and Embarked contain some missing value
```

```python
d=data_train['Survived'].dropna()
```

# 3) Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not. 

```python
data_train['Has_Cabin']=data_train['Cabin'].notnull()
```

```python
data_train.head()
```

# 4) Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc. There should be five titles: Mr, Mrs, Master, Miss, and Special. 

```python
data_train['Title'] = data_train.Name.apply(lambda x: re.search(' ([A-Z][a-z]+)\.', x).group(1))
data_train['Title'] = data_train['Title'].replace({'Mlle':'Miss', 'Mme':'Mrs', 'Ms':'Miss'})
data_train['Title'] = data_train['Title'].replace(['Don', 'Dona', 'Rev', 'Dr',
                                            'Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Special')

```

```python
data_train.head(10)
```

# 5) Impute the missing values using .fillna() method.

```python
new_data_train=data_train.fillna(data_train.mean())
new_data_train.count()
```

# 6) Create two new columns “Categorical_age” and “Categorical_fare”, where you will put people that are within a certain range of age or fare in the same bin. 
You can do this by using the pandas function qcut() to bin your numerical data. 

```python
new_data_train['CatAge'] = pd.qcut(data_train.Age, q=4, labels=False )
new_data_train['CatFare']= pd.qcut(data_train.Fare, q=4, labels=False)
new_data_train.head()
```

# 7) Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.

```python
new2=data_train.drop(['Cabin','Name','PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch'], axis=1)
new2.head()
```

# 8) Transform all variables into numeric ones using get_dummies().

```python
data_dum = pd.get_dummies(new2, drop_first=True)
data_dum
```

```python
n_data=data_dum.fillna(data_dum.mean())
n_data.count()
```

```python
x=n_data[[ 'Pclass', 'Has_Cabin', 'CatAge', 'CatFare', 'Sex_male', 'Embarked_Q', 'Embarked_S', 'Title_Miss', 'Title_Mr', 'Title_Mrs', 'Title_Special']]
y=n_data['Survived']
```

# 9) Now we need training set and testing set. Use train_test_split() to split your dataset. 
X(Features) will be the dataset values without the column “Survived” and y(Targets) will be your “Survived” column values.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# 10) Use logistic regression by passing your X_train and y_train.

```python
from sklearn.linear_model import LogisticRegression
lreg=LogisticRegression()

lreg.fit(x_train,y_train)
```

```python
pred=lreg.predict(x_test)
```

```python
df=pd.DataFrame({'Actual':y_test, 'predict':pred})
df
```

# 11) Print the accuracy score.

```python
score=lreg.score(x_test,y_test)
score
```
